# Replaced Data Values

```python
import numpy as np
df= pd.DataFrame(np.random.randn(7,3))
df.iloc[:4,1] = np.nan
df.iloc[:2,2] = np.nan
print(df)
```

```python
df= df.replace(np.nan, -999)
print(df)
```

```python
df= df.replace(-999, -9)
df
```

```python
df= df.replace([-999, -9], [np.nan, 0])
df
```

# Renaming index using function mapping

```python
data = pd.DataFrame(np.arange(12).reshape((3,4)),
                    index=["ohio","coloroda","newyork"],
                    columns=["one","two","three","four"])
display(data)
```

```python
transform = lambda x: x[:4].upper()
data= data.index.map(transform)
data
```

# Detecting Filtering outliers

```python
data = pd.DataFrame(np.random.randn(100,4))
data.describe()
```

```python
col = data[2]
#col
```

```python
col[np.abs(col) > 3]
```

```python
# to select all values having a value greater than 3 or -3
# you can use "any" method on a boolean dataframe
# any method will check all cells

outliers= data[(np.abs(data) > 3).any(1)]
print(len(outliers), len(data))
outliers
```

```python
# check the difference
data[(np.abs(data) > 3)]
```
