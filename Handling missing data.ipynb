```python
import pandas as pd
from numpy import nan as NA
data= pd.DataFrame([[1.,1.6,3.],[1.,NA,NA],[NA,NA,NA],[NA,6.5,3.]],columns= list("abc") )
display(data)    
```

```python
# remove all rows that have NaN values
cleaned= data.dropna()
cleaned
```

```python
# how ="all"  remove rows only all data is NaN
cleaned= data.dropna(how="all")
display(cleaned)   # display work same as print
```

# Filling Missing Data

```python
import numpy as np
df= pd.DataFrame(np.random.randn(7,3))
df.iloc[:4,1] = NA
df.iloc[:2,2] = NA
display(df)
```

```python
# fillna fill NaN values
df.fillna(1.5)
```

```python
# column no 1 NaN value will be filled 0.5 and column no 2 with fill 0
df= df.fillna({1:0.5, 2:0})
df
```

```python
# fill not values from left column (axis=0 mean from top row)
df= df.fillna(method="ffill")
df.fillna(method="ffill", axis=1)
df.fillna(method="ffill",limit=1, axis=1)
```
